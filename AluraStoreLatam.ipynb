#1. Análisis de facturación



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# 1. Análisis de facturación

# Calcular el ingreso total de cada tienda (suma de la columna Precio)
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Crear un DataFrame resumen
ingresos = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
})

# Mostrar resultados ordenados
print("Ingresos totales por tienda:")
print(ingresos)

# Visualización opcional
import matplotlib.pyplot as plt

plt.bar(ingresos['Tienda'], ingresos['Ingreso Total'], color='skyblue')
plt.title('Ingresos por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total ($)')
plt.show()
```

# 2. Ventas por categoría

```python
# 2. Ventas por categoría

# Unir todas las tiendas en un solo DataFrame
todas_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

# Calcular ventas totales por categoría de producto
ventas_categoria = todas_tiendas.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria.columns = ['Categoría del Producto', 'Ventas Totales']

print("Ventas totales por categoría:")
print(ventas_categoria)

# Visualización
plt.bar(ventas_categoria['Categoría del Producto'], ventas_categoria['Ventas Totales'], color='lightgreen')
plt.title('Ventas por Categoría de Producto')
plt.xticks(rotation=45)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# 3. Calificación promedio de la tienda

calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

calificaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
})

print("Calificación promedio por tienda:")
print(calificaciones)

# Visualización
plt.bar(calificaciones['Tienda'], calificaciones['Calificación Promedio'], color='orange')
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Promedio de Calificación')
plt.show()
```

# 4. Productos más y menos vendidos

```python
# 4. Productos más y menos vendidos

# Unir todas las tiendas
todas_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

# Contar cuántas veces se vendió cada producto
productos_vendidos = todas_tiendas['Producto'].value_counts().reset_index()
productos_vendidos.columns = ['Producto', 'Cantidad de Ventas']

# Producto más vendido
producto_mas_vendido = productos_vendidos.iloc[0]
# Producto menos vendido
producto_menos_vendido = productos_vendidos.iloc[-1]

print("📈 Producto más vendido:")
print(producto_mas_vendido)
print("\n📉 Producto menos vendido:")
print(producto_menos_vendido)

# Visualización (Top 10 más vendidos)
import matplotlib.pyplot as plt

top10 = productos_vendidos.head(10)
plt.barh(top10['Producto'], top10['Cantidad de Ventas'], color='mediumseagreen')
plt.title('Top 10 productos más vendidos')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.gca().invert_yaxis()
plt.show()
```

# 5. Envío promedio por tienda

```python
# 5. Envío promedio por tienda

# Calcular el costo de envío promedio por tienda
envio_prom_t1 = tienda['Costo de envío'].mean()
envio_prom_t2 = tienda2['Costo de envío'].mean()
envio_prom_t3 = tienda3['Costo de envío'].mean()
envio_prom_t4 = tienda4['Costo de envío'].mean()

# Crear DataFrame con resultados
envio_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [envio_prom_t1, envio_prom_t2, envio_prom_t3, envio_prom_t4]
})

print("Costo de envío promedio por tienda:")
print(envio_promedio)

# Visualización
plt.bar(envio_promedio['Tienda'], envio_promedio['Costo de Envío Promedio'], color='cornflowerblue')
plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo promedio de envío ($)')
plt.show()
```

Los productos con mayor demanda se concentraron principalmente en las tiendas 1 y 2.
